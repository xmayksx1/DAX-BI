```python
import pandas as pd
import numpy as np

# 1. Carga de datos
ruta = r"C:\Users\Usuario\OneDrive - Aceitera El Real\Escritorio\Planificación - Planificación y Control\35. Dash Temporal 2026\Prueba de Automatización\Flujos de Trabajo\Detalle del Pedido.xlsx"
df = pd.read_excel(ruta, sheet_name="Hoja2", header=2)

# 2. Selección y limpieza de datos vacíos
pedido = df[['EAN', 'Pedido']].copy().dropna(subset=['EAN', 'Pedido'])

# 3. Crear identificadores de posición
# Grupo: Define en qué bloque horizontal va (0, 1, 2...)
# Fila_Interna: Define la posición vertical dentro del bloque (0 a 14)
pedido['Grupo'] = np.arange(len(pedido)) // 15
pedido['Fila_Interna'] = np.arange(len(pedido)) % 15

# 4. Pivotar para expandir horizontalmente
# El índice serán las 15 filas, las columnas serán los grupos
pedido_horizontal = pedido.pivot(index='Fila_Interna', columns='Grupo', values=['EAN', 'Pedido'])

# 5. Reordenar las columnas para que EAN y Pedido del mismo grupo estén juntos
# Por defecto Pandas agrupa todos los EAN y luego todos los Pedidos.
# Con esto los ponemos: EAN_0, Pedido_0, EAN_1, Pedido_1...
pedido_horizontal = pedido_horizontal.reindex(sorted(pedido_horizontal.columns, key=lambda x: x[1]), axis=1)

# 6. Simplificar nombres de columnas (opcional, para quitar el nivel "Grupo")
pedido_horizontal.columns = [f"{col[0]}_{col[1]}" for col in pedido_horizontal.columns]

# 7. Guardar el resultado final
ruta_salida = r"C:\Users\Usuario\OneDrive - Aceitera El Real\Escritorio\Planificación - Planificación y Control\35. Dash Temporal 2026\Prueba de Automatización\Flujos de Trabajo\Detalle_Pedidos_Horizontal.xlsx"
pedido_horizontal.to_excel(ruta_salida, index=False)

print(f"✅ Proceso completado. Archivo guardado en: {ruta_salida}")
display(pedido_horizontal)
```
