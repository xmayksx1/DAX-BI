```python
import pyodbc
import pandas as pd
import os

# --- CONFIGURACIÓN DE LA CONEXIÓN A SQL SERVER ---
# Se definen los parámetros de acceso al servidor de base de datos
conexion = pyodbc.connect(
    'DRIVER={SQL Server};'
    'SERVER=172.18.1.46;'
    'DATABASE=DWH;'
    'UID=CUBOS;'
    'PWD=#cUb05@c3r3@1'
)

# --- DEFINICIÓN DE LA CONSULTA SQL ---
# Seleccionamos los campos necesarios de la vista de ventas para el periodo específico
query = """
SELECT BarraUnidad, Documento, ZGZona, FECHAFACTURA, FANOMES, FANO, FMES,
       ZGDepartamento, CANALDISTRIBUCION, CodigoCliente, NombreCliente,
       ITMCodigo, ITMDescripcion, CantidadFacturada, Cajas, PesoNeto,
       MontoTotalDOL, Bonificado, CostoTotalDOL, CodigoVendedor
FROM V001_CuboVentaInforLn
WHERE FANOMES IN ('202604')
"""

# --- CONFIGURACIÓN DE RUTAS DE SALIDA ---
# Definimos la carpeta de destino y el nombre del archivo resultante
ruta_destino = r"C:\Users\Usuario\OneDrive - Aceitera El Real\Escritorio\Planificación - Planificación y Control\35. Dash Temporal 2026\Histórico CSV Venta"
nombre_archivo = "Reporte_Ventas_2026.csv"
ruta_completa = os.path.join(ruta_destino, nombre_archivo)


try:
    print("Extrayendo datos de SQL Server...")

    # Extraemos los datos directamente a un DataFrame de Pandas
    df = pd.read_sql(query, conexion)

    # Verificamos si la ruta de destino existe; si no, se podría crear (opcional)
    # En este caso, el script procede a guardar el archivo CSV

    # Exportación del DataFrame a formato CSV

    df.to_csv(ruta_completa, index=False, encoding='utf-8-sig')

    print(f"¡Éxito! El archivo ha sido exportado en: {ruta_completa}")

    # Visualización de las primeras filas en el entorno de Jupyter
    display(df.head())

except Exception as e:
    # Captura de errores durante la extracción o guardado de datos
    print(f"Error al procesar la consulta o exportar: {e}")

finally:
    # CIERRE DE CONEXIÓN: Esencial para liberar recursos en el servidor SQL
    conexion.close()
    print("Conexión cerrada.")
```
